 import the data

```python
df_ratings= pd.read_csv("ratings.csv")
df_movies= pd.read_csv("movies.csv")
```

```python
df_movies.head()
```

We can see that, this dataset contains movieId, the title of the movie, and its genre.

We need a dataset that contains the userId, movie title and its ratings.

We have this information in two different files: ratings.csv and movies.csv.

To get our desired information in a single dataframe, we can merge the two dataframes objects on the movieId column since it is common between the two dataframes.

We can do this using the merge() function from the Pandas library, as shown below.

```python
movie_data = pd.merge(df_ratings, df_movies, on='movieId')
movie_data.head()
```

We can see our newly created dataframe contains userId, title and rating of the movie as required.


Now, it is the time to find the similarity between the movies.

This matrix will have a lot of null values since every movie is not rated by every user.

We will create the matrix of movie titles and corresponding user ratings.


```python
user_movie_rating = movie_data.pivot_table(index='title', columns='userId', values='rating')
user_movie_rating=user_movie_rating.iloc[:,:200]
user_movie_rating.head()
```

now, we will convert all null values with zero using fillna() function

```python
ratings = user_movie_rating.fillna(0,axis=0)
ratings.head(200)
```

importing cosine_similarity from sikit learn to make similarity matrix between ratings of a movie where each column is a movie name and each row contains the rating assigned by a specific user to that movie.

hint: before using similarity matrix we shoud making standraziation to the data 

```python
from sklearn.metrics.pairwise import cosine_similarity 


def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

df_std = ratings.apply(standardize)
similarity_matrix=pd.DataFrame(cosine_similarity(df_std), columns=df_std.index, index=df_std.index)
similarity_matrix.head(200)
```

We know that each column contains all the user ratings for a particular movie.

Now, let's find 10  movies similar for the Toy Story (1995) 


```python
similarMovies= pd.DataFrame()
similarMovies=similarity_matrix["Toy Story (1995)"].sort_values(ascending=False)[1:10]

similarMovies.head(10)
```
